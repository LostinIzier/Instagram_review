```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
df=pd.read_csv("Downloads\instagram.csv\instagram.csv")
df["review_date"]=pd.to_datetime(df["review_date"])
df=df.drop(columns=['review_description'])
#df.set_index('review_date', inplace=True)
df['month_year'] = df['review_date'].dt.to_period('M')
df2=df.drop(columns=['review_date'])
df2=df2.groupby('month_year').apply(np.mean).reset_index()
df2.rename(columns={0:'rating'},inplace=True)
#df2=pd.DataFrame(df2, index('month-year'),columns('reviews'))
df2
```

```python

df2['rating'] = df2['rating'].astype(float)
x=0;
sr=[]
for i in range(len(df2)):
    a = (x * i + df2['rating'].iloc[i]) / (i + 1)
    sr.append(a)
    x = a
sr = pd.Series(sr, index=df2.index)
df2['Cummilative']=sr.values
df2
```

```python
plt.figure(figsize=(10,5))
df2.plot('month_year','Cummilative')
plt.title('Reviews Over Time')
plt.xlabel('Time (Month-Year)')
plt.ylabel('Reviews');
```
